```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
data = {
    'message': [
        'you won lottery',   #spam words to detect whether they are spam are not spam
        'extra income',
        'won cash prize',
        'limited time offer',
        'get rich quick',
        'Buy 1 get 1 free',
        'but lamborghini with 1 rupee',
        'you won 1 crore coupon',
        'get loan of 100 crore today',
        'congratulations you won lottery',
        'do not miss this',
        'remainder',
        'join this zoom meeting at 10am',
        'Upgrade your plan',
        'Your OTP is',
        'Bank Statement',
        'Update your plan',
        'invoice',
        'verify its you',
        'invoice',
        'payment confirmed'
    ],
    'label': [
        'spam','spam','spam','spam','spam','spam','spam','spam','spam','spam','spam',
        'Not Spam','Not Spam','Not Spam','Not Spam','Not Spam','Not Spam','Not Spam','Not Spam','Not Spam','Not Spam'
    ]
}

df = pd.DataFrame(data)

```

```python
df['label']=df['label'].map({'spam':1,'Not Spam':0})
```

```python
vectorizer=CountVectorizer()
x=vectorizer.fit_transform(df['message'])
y=df['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)  #here x is input y is output
```

```python
model =MultinomialNB() #creating model here
model.fit(x_train,y_train)
```

```python
#evaluating model here
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
confusion=confusion_matrix(y_test,y_pred)
```

```python
#giving input and testing here with new data
new_email=["limited time offer"]
new_email_vectorized=vectorizer.transform(new_email)
prediction=model .predict(new_email_vectorized)

if prediction == 1:
  print("The email is classified as spam.")
else:
  print("The email is classified as not spam.")
```
